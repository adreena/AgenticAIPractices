```python
from openai import OpenAI
import os
from IPython.display import Markdown, display
```

```python
client = OpenAI(
    base_url = "https://openai.vocareum.com/v1",
    api_key="")

MODEL = "gpt-4o-mini"

def get_completion(system_prompt):
    messages = [
        {"role": "system", "content": system_prompt}
        # {"role": "user", "content": user_prompt}
    ]
    try:
        response = client.chat.completions.create(
            model = MODEL,
            messages = messages,
            temperature = 0.7
        )
        return response.choices[0].message.content
    except Exception as err:
        return f"error occrued: {err}"

def display_responses(response):
    """Helper function to display responses as Markdown, horizontally."""
    markdown_string = "<table><tr>"
    markdown_string += f"<td>Response:<br />{response}</td>"
    markdown_string += "</tr></table>"
    display(Markdown(markdown_string))
```

```python
prompt = f"""
you are a programming assistant.
I need a python script to write a DFS function that receives input of queue and implements Depth-first-Search. provde step-by-step outline for this script.
"""


response  = get_completion(prompt)
```

```python
display_responses(response)
```

```python
gate_checker_prompt = f"""
you are a python script validator.
Validate this script {response} and show me 2 test cases that pass through the code without error. if the code has error respond with false; and if the code is correct respond with true
"""

gate_check_response = get_completion(gate_checker_prompt)
display_responses(gate_check_response)
```
